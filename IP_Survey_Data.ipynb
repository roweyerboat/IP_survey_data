```python
# Since there are 7 different options to choose from ranging from "Strongly Disagree" to "Strongly Agree" 
# We can replace each of those choices with a number 1-7 and then the nans become 0s so we can do some computations
replacer = {'Strongly Disagree': 1, 'Disagree': 2, 'Somewhat Disagree': 3, "Neutral": 4, "Somewhat Agree": 5, "Agree": 6, "Strongly Agree": 7}
```

```python
survey_data['Clear about change we seek'] = survey_data['Clear about change we seek'].replace(replacer)
survey_data['Clarity direction of work ahead'] = survey_data['Clarity direction of work ahead'].replace(replacer)
survey_data['Clear about roles and responsibilities'] = survey_data['Clear about roles and responsibilities'].replace(replacer)
survey_data['Confident plan will achieve goals'] = survey_data['Confident plan will achieve goals'].replace(replacer)
survey_data['Feel more equipped'] = survey_data['Feel more equipped'].replace(replacer)
survey_data['Equipped and supported'] = survey_data['Equipped and supported'].replace(replacer)
survey_data['Understand state of instruction'] = survey_data['Understand state of instruction'].replace(replacer)
survey_data['Valuable use of my time:'] = survey_data['Valuable use of my time:'].replace(replacer)
survey_data['I learned something:'] = survey_data['I learned something:'].replace(replacer)
```
